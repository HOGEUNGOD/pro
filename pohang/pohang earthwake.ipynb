```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.font_manager import FontProperties

magnitude_excel = np.array(pd.read_excel('magnitude.xlsx'))
magnitude_x = magnitude_excel[:,2]*8/2001
magnitude_y = magnitude_excel[:,3]*8/938-2

net_injection = np.array(pd.read_excel('net injection.xlsx'))
net_injection_x = net_injection[:,2]*8/2001
net_injection_y=  net_injection[:,3]*1000/938

injection = np.array(pd.read_excel('injection.xlsx'))
injection_x = injection[:,2]*8/2001
injection_y=  injection[:,3]*1000/938

back_flow = np.array(pd.read_excel('back flow.xlsx'))
back_flow_x = back_flow[:,2]*8/2001
back_flow_y=  back_flow[:,3]*1000/938


fig, ax1 = plt.subplots(figsize=[10, 5])
ax = ax1.twinx()

ax.scatter(magnitude_x, magnitude_y, linewidths=2.5,facecolors ='none', edgecolors='r',label='Magnitude',zorder=1)

ax.set_ylim(-2,6)
ax.set_xlim(0,8)

ax1.plot(injection_x, injection_y, c='blue' , label='Injection',linewidth='2.5', zorder=1)
ax.plot([-1],[-1],c= 'blue', label='Injection')
ax1.plot(net_injection_x, net_injection_y,c= 'black', label='Net injection',linewidth='2.5', zorder=2)
ax.plot([-1],[-1],c= 'black', label='Net injection')
ax1.plot(back_flow_x, back_flow_y,c= 'gold', label='Net injection',linewidth='2.5', zorder=2)
ax.plot([-1],[-1],c= 'gold', label='Flow back')


ax1.set_ylim(0,1000)
labels = [item.get_text() for item in ax.get_xticklabels()]
labels[0]= '23 Oct 2015'
labels[1]= '31 Jan 2016'
labels[2]= '10 May 2016'
labels[3]= '18 Aug 2016'
labels[4]= '26 Nov 2016'
labels[5]= '06 Mar 2017'
labels[6]= '14 Jun 2017'
labels[7]= '22 Sep 2017'
labels[8]= '31 Dec 2017'

ax.legend()
ax.set_ylabel('Magnitude '+r'( $M_{L}$ )')
ax1.set_ylabel('Injection volume' + r'( $ m^{3}/day $ )' +'\n'+ 'Net cumulative injection volume' + r'( $ \times10m^{3} $ )')


font = FontProperties()
# font.set_family('serif')
# font.set_name('Times New Roman')
# font.set_style('italic')
ax.tick_params(direction='in')
ax1.tick_params(direction='in')
ax1.set_xticklabels(labels,fontsize=9, ha="center")
ax1.tick_params(axis='x', pad=10, length=5)
ax1.grid(axis='y')
ax1.tick_params(zorder=10)
ax.set_axisbelow(True)
ax1.set_axisbelow(True)
# ax1.xaxis.tick_bottom()
ax1.text(0.75,440,'Stage 1\n  PX-2')
ax1.text(3.95,900,'Stage 2\n  PX-1')
ax1.text(5.1, 690, 'Stage 3\n  PX-2')
ax1.text(6.205, 700, 'Stage 4\n  PX-1')
ax1.text(6.805, 695, 'Stage 5\n  PX-2')


ax1.patch.set_visible(False)


plt.savefig('./abc.jpeg', dpi=300)
plt.show()
```
