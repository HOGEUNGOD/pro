```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img = cv2.imread('../p12.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
g = float(0.4)
out = gray.astype(np.float)
out = ((out / 255) ** (1 / g)) * 255
out = out.astype(np.uint8)
plt.imshow(out,cmap='gray')
plt.axis('off')
plt.savefig('../data/off.png',dpi=300,bbox_inches='tight')

plt.show()
plt.close()

```
